#### `students.append(student)`

We now take that dictionary and **add it to our main list called `students`**.

---

````python
def add_student():
    name = input("Enter student name: ")
    roll = input("Enter roll number: ")
    marks = []
    for sub in subjects:
        while True:
            try:
                score = float(input(f"Enter marks for {sub}: "))
                if 0 <= score <= 100:
                    marks.append(score)
                    break
                else:
                    print("Marks must be between 0 and 100.")
            except ValueError:
                print("Please enter a valid number.")
    student = {"name": name, "roll": roll, "marks": marks}
    students.append(student)
    print(f"Student {name} added successfully!\n")
````

### Calculate Result

* loop over all students.
* calculate total marks and percentage.
* decide the grade using if-else conditions.
* add all this info into the student’s data.

---

#### `def calculate_results():`

This is where we **define a new function** called `calculate_results`.

When we run this function, it will:

> Go through every student in the list and calculate:
>
> * Total marks
> * Percentage
> * Grade

---

#### `for student in students:`

This starts a loop that goes through every student in our `students` list.

If you’ve added 3 students, this loop will run 3 times — once for each student dictionary.

---

#### `total = sum(student['marks'])`

This line adds up all the marks the student got.

Example:

If a student’s marks are `[85, 90, 78]`,
then `total = 85 + 90 + 78 = 253`

We’re using the `sum()` function to quickly add everything in the `marks` list.

---

#### `percentage = total / len(subjects)`

This line calculates the **percentage**.

How?

* It divides the `total` by the number of subjects.
* `len(subjects)` tells us how many subjects there are — like 3 if we have Math, Science, English.

So for a total of `253` and 3 subjects:
`percentage = 253 / 3 = 84.33`

The result is stored in a variable called `percentage`.

---

#### `student['total'] = total`

We now **add a new field** called `'total'` into the student’s dictionary.

This lets us keep the total marks along with the name, roll, and marks.

So the student dictionary grows like this:

```python
{
  "name": "Amit",
  "roll": "R001",
  "marks": [85, 90, 78],
  "total": 253
}
```

---

#### `student['percentage'] = percentage`

Same idea here — we store the calculated percentage in the student’s dictionary.

So now it becomes:

```python
"percentage": 84.33
```

This makes it easy to display or sort students later based on percentage.

---

#### `if percentage >= 90:`

Now we start checking the **grading rules**.

This one says:

> “If the student got 90% or more, give them an A+”

---

#### `grade = 'A+'`

We assign `'A+'` to a variable called `grade`.

---

#### `elif percentage >= 75:`

Else if the student didn’t get 90+, but got 75 or more, give them an A.

This pattern continues:

```python
elif percentage >= 60: grade = 'B'
elif percentage >= 40: grade = 'C'
else: grade = 'F'
```

So the logic is:

* 90+ → A+
* 75–89 → A
* 60–74 → B
* 40–59 → C
* Below 40 → F (fail)

---

#### `student['grade'] = grade`

Finally, we store the grade in the student dictionary.

At the end, the full student dictionary looks something like this:

```python
{
  "name": "Amit",
  "roll": "R001",
  "marks": [85, 90, 78],
  "total": 253,
  "percentage": 84.33,
  "grade": "A"
}
```

````python
def calculate_results():
    for student in students:
        total = sum(student['marks'])
        percentage = total / len(subjects)
        student['total'] = total
        student['percentage'] = percentage
        if percentage >= 90:
            grade = 'A+'
        elif percentage >= 75:
            grade = 'A'
        elif percentage >= 60:
            grade = 'B'
        elif percentage >= 40:
            grade = 'C'
        else:
            grade = 'F'
        student['grade'] = grade
````

### Find Topper

* We check if there are any students.
* If there are, we look through all of them.
* We find the one with the **highest percentage**.
* Then we proudly show their name and score — like announcing a winner

---

#### `def find_topper():`

This is where we define a new function called `find_topper`.

It **finds the student with the highest percentage** and prints their name and score.

Later, we can run it by simply calling:

```python
find_topper()
```

---

#### `if not students: return`

First, we do a simple check.

* `students` is our list of student dictionaries.
* `if not students:` means:

> “If the `students` list is **empty**, then stop right here.”

Why do we do this?

To **avoid errors**.

If there are no students, there's no topper to find — so we exit the function.

---

#### `topper = max(students, key=lambda x: x['percentage'])`

This is the most important line — we are finding the **topper** here.

Let’s break it down:

* `max(students, ...)`: This looks through the list of students and picks the one with the **maximum** value.
* `key=lambda x: x['percentage']`: This tells Python:

> “Look at each student’s `'percentage'` and compare based on that.”

> "Find the student in the list who has the highest percentage."

The result — the student with the highest percentage — is stored in the variable `topper`.

Example:

If we have three students with percentages:

* Asha – 82.5%
* Ravi – 91.3%
* Meena – 76.0%

This line will pick Ravi's dictionary and store it in `topper`.

---

#### `print(f"\n Topper: {topper['name']} ({topper['percentage']:.2f}%)")`

This line **prints the name and percentage** of the topper.

* `topper['name']`: gets the name from the dictionary
* `topper['percentage']:.2f`: prints the percentage with 2 decimal places
* `\n`: adds a blank line before the message (for clean formatting)

Example output:

```
Topper: Ravi (91.30%)
```

````python
def find_topper():
    if not students: return
    topper = max(students, key=lambda x: x['percentage'])
    print(f"\n Topper: {topper['name']} ({topper['percentage']:.2f}%)")
````

### Subject Analysis

* We collect all students’ marks into a NumPy array.
* Then, for each subject:

  * We extract the marks.
  * We calculate the average, highest, and lowest score.
  * And we print that in a clean summary.

---

#### `def subject_analysis():`

This line is defining a function named `subject_analysis`.

> It will **analyze marks for each subject** across all students and tell us:

* The **average score**
* The **highest score**
* The **lowest score**

---

#### `print("\n Subject-wise Analysis:")`

This just prints a heading.

* `\n` puts it on a **new line** (for better formatting).
* The message tells the user that we’re about to do **subject-wise analysis**.

Example output:

```
Subject-wise Analysis:
```

---

#### `mark_array = np.array([s['marks'] for s in students])`

This is a very important line where **NumPy** comes into action.

Let’s break it down:

* `[s['marks'] for s in students]` is a list comprehension. It collects the **marks list of every student**.

  * Example:

    ```python
    [
      [85, 90, 78],
      [88, 92, 80],
      [76, 70, 60]
    ]
    ```

* `np.array(...)` converts this into a **NumPy array**, which looks like a table (rows and columns), like this:

  ```
  [[85 90 78]
   [88 92 80]
   [76 70 60]]
  ```

Now we can easily calculate column-wise values — like average for Math, Science, etc.

---

#### `for i, sub in enumerate(subjects):`

This line starts a **loop** that lets us go through each subject **with its index**.

* `enumerate(subjects)` gives both the **index** (`i`) and the **subject name** (`sub`).

  * Example:

    * `i = 0`, `sub = "Math"`
    * `i = 1`, `sub = "Science"`
    * `i = 2`, `sub = "English"`

This loop helps us go **column by column** in the NumPy array, matching each one to a subject name.

---

#### `sub_marks = mark_array[:, i]`

This line is slicing the NumPy array.

* `mark_array[:, i]` means:

> “Give me **all rows** (`:`) in the **i-th column**”

Each column represents one subject. So:

* `i = 0`: marks for Math
* `i = 1`: marks for Science
* `i = 2`: marks for English

So `sub_marks` will be something like:

```python
array([85, 88, 76])
```

→ marks of all students in that one subject.

---

#### `print(f"{sub} - Avg: {np.mean(sub_marks):.2f}, Max: {np.max(sub_marks)}, Min: {np.min(sub_marks)}")`

This prints the **summary for each subject**:

* `sub`: subject name (like "Math")
* `np.mean(sub_marks)`: average marks
* `np.max(sub_marks)`: highest marks
* `np.min(sub_marks)`: lowest marks
* `:.2f`: makes the average look nice (2 decimal places)

Example output:

```
Math - Avg: 83.00, Max: 88, Min: 76
Science - Avg: 84.00, Max: 92, Min: 70
English - Avg: 72.67, Max: 80, Min: 60
```



````python
def subject_analysis():
    print("\n Subject-wise Analysis:")
    mark_array = np.array([s['marks'] for s in students])
    for i, sub in enumerate(subjects):
        sub_marks = mark_array[:, i]
        print(f"{sub} - Avg: {np.mean(sub_marks):.2f}, Max: {np.max(sub_marks)}, Min: {np.min(sub_marks)}")
````

### Update Pass/Fail

* We go through each student.
* We check if they got **less than 40** in any subject.
* If yes: we mark them as **Fail** and list the failed subjects.
* If not: we mark them as **Pass**.

This function helps us later when we want to display only **failed students**, or show results with status and subject-wise remarks.

---

#### `def update_pass_fail():`

We are defining a new function called `update_pass_fail`.

> It checks each student’s marks and decides:
>
> * Whether they **Passed** or **Failed**
> * If failed, **which subjects** they failed in

We’ll run this later using:

```python
update_pass_fail()
```

---

#### `for s in students:`

This line starts a loop that goes through every student in the list.

* `students` is our main list that holds all student data.
* Each `s` is one student’s dictionary, like:

```python
{
  "name": "Amit",
  "roll": "R001",
  "marks": [85, 90, 32],
  ...
}
```

---

#### `failed_subjects = [subjects[i] for i, m in enumerate(s['marks']) if m < 40]`

This line uses a **list comprehension** to find out which subjects the student failed.

Let’s understand this step by step:

* `enumerate(s['marks'])`:
  This gives you both the index (`i`) and the mark (`m`) for each subject.
  Example:

  ```python
  [(0, 85), (1, 90), (2, 32)]
  ```

* `if m < 40`:
  We only want the subjects where the marks are **less than 40**.

* `subjects[i]`:
  This gives us the **subject name** (like "Math", "Science", etc.) using the index.

So if a student has:

```python
marks = [85, 90, 32]
```

This line will return:

```python
failed_subjects = ['English']
```

---

#### `if failed_subjects:`

This checks:

> “Does the student have any failed subjects?”

If yes → we update their status to **"Fail"**

If no → they passed in all subjects

---

#### `s['status'] = 'Fail'`

If the student has failed at least one subject, we add a new key to their dictionary:

```python
"status": "Fail"
```

Now we can easily filter failed students later.

---

#### `s['failed_subjects'] = failed_subjects`

We store the **list of subjects** they failed in.

Example:

```python
"failed_subjects": ["English"]
```

---

#### `else:`

This runs **only if the student didn’t fail any subject**.

---

#### `s['status'] = 'Pass'`

We mark the student as having passed.

---

#### `s['failed_subjects'] = []`

Even though the student passed, we still add `"failed_subjects"` as an **empty list** so the structure is the same for everyone.

````python
def update_pass_fail():
    for s in students:
        failed_subjects = [subjects[i] for i, m in enumerate(s['marks']) if m < 40]
        if failed_subjects:
            s['status'] = 'Fail'
            s['failed_subjects'] = failed_subjects
        else:
            s['status'] = 'Pass'
            s['failed_subjects'] = []
````

### List All Failed Students

* We go through each student one by one.
* If they failed, we print their name, roll number, and the subject(s) they failed in.
* This function is great when you want to **quickly identify which students need support**.

---

#### `def list_failed_students():`

We are creating a function named `list_failed_students`.

> To go through the students list and **print the names of those who failed**, along with the **subjects they failed in**.

Once this function is defined, we can run it by simply calling:

```python
list_failed_students()
```

---

#### `print("\n Failed Students:")`

This line just prints a heading to let us know what we’re about to see.

* `\n` adds a blank line before it — to separate it nicely from previous outputs.
* The message `"Failed Students:"` will look like a title.

Example:

```
Failed Students:
```

---

#### `for s in students:`

This starts a loop to go through each student in the list called `students`.

Each `s` is one student's dictionary. For example:

```python
{
  "name": "Amit",
  "roll": "R001",
  "marks": [85, 90, 32],
  "status": "Fail",
  "failed_subjects": ["English"]
}
```

We're going to check if this student failed.

---

#### `if s.get('status') == 'Fail':`

Here’s where we check the student's pass/fail status.

* `s.get('status')` looks for the `'status'` key in the student's dictionary.
* If the value is `"Fail"`, it means the student failed.

We use `.get()` instead of `s['status']` just in case `'status'` doesn't exist yet — that way, we avoid errors.

---

#### `print(f"{s['name']} (Roll: {s['roll']}) - Failed in: {', '.join(s['failed_subjects'])}")`

This line prints the actual result in a **clean and clear format**.

Let’s break it down:

* `s['name']`: student’s name
* `s['roll']`: roll number
* `s['failed_subjects']`: this is a list like `["Math", "English"]`

`', '.join(...)` turns that list into a string like:

```
Math, English
```

So the final print might look like:

```
Amit (Roll: R001) - Failed in: English
```

````python
def list_failed_students():
    print("\n Failed Students:")
    for s in students:
        if s.get('status') == 'Fail':
            print(f"{s['name']} (Roll: {s['roll']}) - Failed in: {', '.join(s['failed_subjects'])}")
````

### Class Average, Min, Max Total

* We collect all total marks from every student.
* We calculate and print:

  * The average total
  * The highest total
  * The lowest total

This gives us a quick snapshot of how the class is doing overall — like a mini report card for the whole class.

---

#### `def class_statistics():`

We are **defining a new function** named `class_statistics`.

> To show a quick summary of the class performance — including:

* Class average (total marks)
* Highest total score

#### `def load_from_csv(filename="students.csv"):`

#### `reader = csv.DictReader(f)`

This creates a **CSV reader** that reads each row of the file as a **dictionary** — where the column names become keys.

So if a CSV row looks like:

#### `for row in reader:`

#### `marks = [float(row[sub]) for sub in subjects]`

#### The block inside the loop:

```python
student = {
    "name": row["Name"],
    "roll": row["Roll"],
    "marks": marks,
    "total": float(row["Total"]),

    "status": row.get("Status", "")
}
```

We now create a student dictionary from the CSV row — just like we do when adding a student manually.

Key notes:

* We convert `"Total"` and `"Percentage"` to `float` (because they were stored as text).

Now we have the student data back in the right format.

#### `students.append(student)`

We add the newly created student dictionary to the `students` list.

Now we’re rebuilding our data in memory, row by row.

````python
def load_from_csv(filename="students.csv"):
    students.clear()
    with open(filename, "r") as f:
        reader = csv.DictReader(f)
        for row in reader:
            marks = [float(row[sub]) for sub in subjects]
            student = {
                "name": row["Name"],
                "roll": row["Roll"],
                "marks": marks,
                "total": float(row["Total"]),
                "percentage": float(row["Percentage"]),
                "grade": row["Grade"],
                "status": row.get("Status", "")
            }
            students.append(student)
    print(f"Data loaded from {filename}")
````

#### `from collections import Counter`

```python
Counter(['A', 'B', 'A', 'C']) → {'A': 2, 'B': 1, 'C': 1}
```

#### `def grade_distribution():`

This line defines a function named `grade_distribution`.

Purpose of the function?

> To go through all students and tell us how many got grade A, B, C, and so on.

```python
grade_distribution()
```

#### `grades = [s['grade'] for s in students]`

#### Subject-Wise Toppers

* For each subject, we find the **highest mark**.
* Then we find the **student(s)** who got that mark.
* We print their names as the subject toppers.

---

#### `def subject_wise_toppers():`

This line defines a new function called `subject_wise_toppers`.

> To find out **who scored the highest marks in each subject**, and then print the name(s) of those toppers.

So if you want to know:

* Who topped in Math?
* Who topped in Science?
* Who topped in English?

This function answers all that.

---

#### `if not students: return`

This is a safety check.

> It checks whether there are **any students** in the `students` list.

* If the list is empty (no students added yet), it just stops and returns — no need to go further.

This avoids crashing the code or printing nonsense.

---

#### `marks_array = np.array([s['marks'] for s in students])`

This is where NumPy comes in to help.

Let’s understand this:

* `[s['marks'] for s in students]` collects just the marks of all students.
  Example:

  ```python
  [
    [85, 90, 78],
    [88, 91, 92],
    [85, 91, 78]
  ]
  ```

* `np.array(...)` converts this into a NumPy array:

  ```
  [[85 90 78]
   [88 91 92]
   [85 91 78]]
  ```

This makes it easier to work with each subject’s marks column-wise.

---

#### `for i, sub in enumerate(subjects):`

This loop lets us go through each subject **along with its index**.

* `enumerate(subjects)` gives:

  * `i = 0`, `sub = "Math"`
  * `i = 1`, `sub = "Science"`
  * `i = 2`, `sub = "English"`

This helps us work column-by-column in our NumPy marks array.

---

#### `max_score = np.max(marks_array[:, i])`

This line finds the **highest mark** in the `i-th` subject.

* `marks_array[:, i]` means:

  > “Give me all students’ marks for subject number `i`”

Example:

* For Science (`i=1`), it might give `[90, 91, 91]`
* `np.max(...)` then gives: `91`

So now we know the highest mark in that subject.

---

#### `toppers = [s['name'] for s in students if s['marks'][i] == max_score]`

This line finds **who got the top score** in this subject.

Let’s break it down:

* `for s in students`: Go through each student.
* `s['marks'][i] == max_score`: Check if their mark for subject `i` equals the highest score.
* `s['name']`: If yes, include the student's name in the toppers list.

This creates a list of all students who topped that subject (handles ties too).

Example:

```python
['Amit', 'Neha']  # Both scored 91 in Science
```

---

#### `print(f"Topper in {sub}: {', '.join(toppers)} ({max_score})")`

This line prints the result in a nice format.

Let’s say:

* Subject: `"Science"`
* Toppers: `['Amit', 'Neha']`
* Max score: `91`

Then it prints:

```
Topper in Science: Amit, Neha (91)
```


````python
def subject_wise_toppers():
    if not students: return
    marks_array = np.array([s['marks'] for s in students])
    for i, sub in enumerate(subjects):
        max_score = np.max(marks_array[:, i])
        toppers = [s['name'] for s in students if s['marks'][i] == max_score]
        print(f"Topper in {sub}: {', '.join(toppers)} ({max_score})")
````
