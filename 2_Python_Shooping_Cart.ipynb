### Step 4: Add to Cart Module

#### `def add_to_cart():`

> We’re creating a function called `add_to_cart`.

> This function will run when someone clicks an "Add to Cart" button in the app.

> Its job is to take the selected product(s) and put them in the shopping cart.

---

#### `selected = product_box.selection()`

> We’re asking the product list (called `product_box`) to tell us **which items the user has selected**.

> It gives back a list of IDs (internal row identifiers) of the selected products.

---

#### `if not selected:`

> This means: “If the user hasn’t selected anything…”

---

> …then we pop up a warning box that says:

---

#### `return`

> This line stops the function right here.

> Since nothing was selected, there’s no point in continuing.

---

#### `for item in selected:`

> Now we go through each selected product one by one.

> This loop will run once for each selected row in the product list.

---

#### `item_data = product_box.item(item, "values")`

> This line gets the **actual data** (like name, category, price) from the selected row.

> It pulls out the values stored in that item.

---

#### `product = {`

> We now make a small dictionary — like a mini info card — for each product.

> It will have three key pieces of info: `name`, `category`, and `price`.

---

#### `"name": item_data[0],`

> We’re saying: the name of the product is the first value in the row.

---

#### `"category": item_data[1],`

> The category (like Electronics, Clothing, etc.) is the second value.

---

#### `"price": float(item_data[2])`

> The price is the third value — we’re converting it to a **float** (a number with decimal) to make sure we can do calculations later, like totals.

---

#### `cart.append(product)`

> We take that product dictionary and **add it to the `cart` list** — just like dropping an item into your shopping cart.

---

### Step 5: Refresh Cart Function

This function **clears the old cart display**, **fills it with the current cart items**, and **shows the updated total amount**.

---

#### `def refresh_cart():`

> We’re defining a function named **`refresh_cart`**.

> This function’s job is to **update the shopping cart on the screen** — both the list of products and the total amount.

---

#### `for item in cart_box.get_children():`

> This line means:

> "Look at all the rows (items) currently shown in the cart display box (called `cart_box`)."

> It gives us a list of all the items currently **visible** in the cart area.

---

#### `cart_box.delete(item)`

> We go one by one and **delete** each item from the cart display.

> In simple terms: we're **clearing the old list** so we can show a fresh, updated one.

---

#### `for product in cart:`

> Now we’re going through all the actual products **saved in the `cart` list** — the real cart data stored in memory.

> We'll use this to **refill the cart display** with the current items.

---

#### `cart_box.insert("", "end", values=(product['name'], product['category'], f"Rs. {product['price']:.2f}"))`

> This line **adds one product at a time** to the cart display on the screen.

> Here’s what it’s doing:
>
> * `""` – we’re not placing the row inside another row (just at the root).
> * `"end"` – add this row to the **end** of the list.
> * `values=(...)` – this is the actual info we want to show in that row:
>
>   * `product['name']` – the product's name (like "Laptop")
>   * `product['category']` – the type of product (like "Electronics")
>   * `f"Rs. {product['price']:.2f}"` – this shows the price with a dollar sign and **2 decimal places**, like `Rs. 10.00`

---

#### `total_label.config(text=f"Total: Rs. {calculate_total():.2f}")`

> Lastly, we **update the total price label** on the screen.

> * We call a function `calculate_total()` that adds up all the prices in the cart.
> * We format the result nicely using `f"Total: Rs..."` so it shows something like:
>   **Total: Rs. 78.50**
> * `.config()` is how we tell the label to show this new text.

````python
def refresh_cart():
    for item in cart_box.get_children():
        cart_box.delete(item)
    for product in cart:
        cart_box.insert("", "end", values=(product['name'], product['category'], f"Rs. {product['price']:.2f}"))
    total_label.config(text=f"Total: Rs. {calculate_total():.2f}")
````

### Step 6: Calculate Total Function

---

#### `def calculate_total():`

> We're creating a function called **`calculate_total`**.
> This function’s job is to **add up all the prices** of the products currently in the shopping cart.

---

#### `return sum(p['price'] for p in cart)`

> Let’s break this down step by step:

* `for p in cart` → This means: go through each product (`p`) in the `cart` list.

* `p['price']` → For each product, take out just the **price**.

* `sum(...)` → This adds up all the prices.

> So this line is basically saying:
> **“Add up the price of every product in the cart and give me the total.”**

> The `return` part means: give back that total wherever this function is used.

---

**Example:**
If the cart has three products priced Rs. 10, Rs. 15, and Rs. 5, this function will return **Rs. 30**.


````python
def calculate_total():
    return sum(p['price'] for p in cart)
````

### Step 7: Sort Cart Function

This function sorts the shopping cart based on either **price** or **category**, and then updates the screen to show the new order.

---

#### `def sort_cart(by):`

> We’re creating a function named **`sort_cart`**.

> This function will **sort the items in the shopping cart**, based on how the user wants to sort it.

> The part inside the brackets — `by` — is like a question we’re asking:

> "What do you want to sort by?" (price or category?)

---

#### `if by == 'price':`

> We're checking:

> **"Did the user ask to sort the cart by price?"**

> If yes, we go ahead and sort the cart that way.

---

#### `cart.sort(key=lambda x: x['price'])`

> This is the line that actually **sorts the `cart` list** by price — from lowest to highest.

> Here's a plain-English breakdown:
>
> * `cart.sort(...)` → Sort the cart in place (no new list created).
> * `key=lambda x: x['price']` → For each item `x` in the cart, look at the **`price`** to decide how to sort it.

> So, it’s saying:
>
> **“Sort the cart by each product's price.”**

---

#### `elif by == 'category':`

> If the user didn’t ask to sort by price, we check:
>
> **“Did they ask to sort by category?”**

---

#### `cart.sort(key=lambda x: x['category'])`

> If yes, we now sort the cart by **product category** (like "Electronics", "Kitchen", etc.).

> Same logic as before, but now we're looking at `x['category']` instead of price.

---

#### `refresh_cart()`

> After sorting is done, we **update the cart display on the screen** using the `refresh_cart()` function — so the user can see the sorted list right away.

````python
def sort_cart(by):
    if by == 'price':
        cart.sort(key=lambda x: x['price'])
    elif by == 'category':
        cart.sort(key=lambda x: x['category'])
    refresh_cart()
````

### Step 8: Export Bill Function

This function **saves the cart to a CSV file** that looks like a receipt, but only if the cart isn't empty. It walks the user through the file-saving process and gives feedback the whole way.

---

#### `def export_bill():`

> We’re creating a function called **`export_bill`**.
> Its job is to **save the shopping cart as a `.csv` file** — like a mini bill or receipt that the user can download.

---

#### `if not cart:`

> This checks:
>
> **"Is the cart empty?"**

> `not cart` means there are no items in the list.

---

#### `messagebox.showwarning("Empty Cart", "No items in cart to export")`

> If the cart is empty, we pop up a warning box with this message:
>
> **Title:** "Empty Cart"
>
> **Message:** "No items in cart to export"

> This is just a friendly way to let the user know there's nothing to save.

---

#### `return`

> We stop the function right here.
>
> No point continuing if the cart is empty.

---

#### `file = filedialog.asksaveasfilename(defaultextension=".csv", filetypes=[("CSV Files", "*.csv")])`

> This line opens a **Save File window** and asks the user **where they want to save the bill**.

> * `defaultextension=".csv"` → If the user doesn't type a file extension, `.csv` will be added automatically.
> * `filetypes=[("CSV Files", "*.csv")]` → It only shows `.csv` files in the save dialog, and saves it in that format.

> Whatever filename the user picks is stored in the variable `file`.

---

#### `if file:`

> This checks:
>
> **"Did the user actually choose a file name?"**

> If they clicked **Cancel**, this will be skipped. If they chose a file, we continue.

---

#### `with open(file, "w", newline="") as f:`

> Now we’re **opening the chosen file** for writing (saving).

> * `"w"` means write mode — we're creating or overwriting the file.
> * `newline=""` makes sure there are no extra blank lines in the CSV.
> * `as f` means we’ll refer to this file as `f` in the lines below.

> Using `with` is nice because it automatically closes the file afterward.

---

#### `writer = csv.writer(f)`

> We’re creating a **CSV writer object** using Python’s `csv` module.
>
> This will help us write rows into the CSV file easily.

---

#### `writer.writerow(["Name", "Category", "Price"])`

> First, we write the **header row** in the file — the column titles.

> So the first line in the file will be:
>
> `Name,Category,Price`

---

#### `for item in cart:`

> Now we go through each item (product) in the shopping cart.

---

#### `writer.writerow([item['name'], item['category'], item['price']])`

> For each product, we write a new line in the file with:
>
> * Product name
> * Product category
> * Product price

> So your CSV file might end up looking like:

```
Name,Category,Price
Laptop,Electronics,800
Book,Stationery,10
```

---

#### `messagebox.showinfo("Exported", f"Bill saved as {file}")`

> After saving is done, we show a friendly message box that says:
>
> **"Exported"**
>
> **"Bill saved as (filename)"**

> It lets the user know that the bill was saved successfully.

````python
def export_bill():
    if not cart:
        messagebox.showwarning("Empty Cart", "No items in cart to export")
        return
    file = filedialog.asksaveasfilename(defaultextension=".csv", filetypes=[("CSV Files", "*.csv")])
    if file:
        with open(file, "w", newline="") as f:
            writer = csv.writer(f)
            writer.writerow(["Name", "Category", "Price"])
            for item in cart:
                writer.writerow([item['name'], item['category'], item['price']])
        messagebox.showinfo("Exported", f"Bill saved as {file}")
````

### Step 9: Function Display Product

---

#### `def display_products(product_list):`

> We are now **creating a small helper function** called `display_products`.

> Its job is simple:
> **Take a list of products and show them inside the product table.**

---

#### `for row in product_box.get_children(): product_box.delete(row)`

> * `product_box.get_children()` gives us all the rows currently in the table.

### Step 10: Search Product Function

This function:

1. Gets the user’s search input.
2. Looks for matches in the product names and categories.
3. Shows the results on the screen.

---

#### `def search_product():`

> We're creating a function called **`search_product`**.
>
> This function is used to **search through the list of products** when a user types something in a search box and clicks "Search."

---

#### `query = search_entry.get().lower()`

> Let’s break this down:

* `search_entry.get()` means:
  "Get whatever the user typed into the search box."

* `.lower()` converts the text to **lowercase** — this is so we don’t worry about capital letters (so “LAPTOP” and “laptop” both work the same).

> We store that search word or phrase in a variable called `query`.

---

#### `results = [p for p in products if query in p['name'].lower() or query in p['category'].lower()]`

> This is the **main search logic**, and it might look a little scary at first — but I’ll explain it like a story:

> Think of it like saying:
>
> * "Go through each product `p` in the list called `products`..."
> * "...and **if** the search word (`query`) is found **inside the product's name** or **category**..."
> * "...then keep that product in a new list called `results`."

> The `.lower()` on both `p['name']` and `p['category']` makes sure we’re matching without worrying about capital letters.

**Example:**
If someone types “elec”, it will match products where the name or category contains “elec” — like “Electronics”.

---

#### `display_products(results)`

> Now that we have a list of matching products, we **send it to another function** (probably written elsewhere) called `display_products`.

> This function will likely **show those search results** on the screen — maybe in a table or list box.

````python
def search_product():
    query = search_entry.get().lower()
    results = [p for p in products if query in p['name'].lower() or query in p['category'].lower()]
    display_products(results)
````

### Step 11: Apply Discount Function

This function:

1. Gets the discount from the user.
2. Applies it to every item in the cart (if it's valid).
3. Updates the prices and shows them on screen.
4. If the input is wrong, it shows a helpful error message instead of crashing.

---

#### `def apply_discount():`

> We're creating a function called **`apply_discount`**.
>
> This function will run when the user types a discount percentage (like 10%) and clicks a button to apply it to everything in the cart.

---

#### `discount = discount_entry.get()`

> We’re **getting the text** that the user typed into the **discount entry box** (like a text field in the GUI).
>
> For example, if the user typed `15`, this line saves that as the variable `discount`.

---

#### `try:`

> This starts a **try block**, which means:
>
> “Let’s try the next few lines, but if something goes wrong (like bad input), we’ll handle it.”

> It helps the program avoid crashing if the user enters something weird, like letters instead of numbers.

---

#### `rate = float(discount)`

> We take the text the user typed (`discount`, like `"10"` or `"15"`), and **convert it into a number** using `float()`.
>
> So if the user typed `"10"`, this becomes the number `10.0`.

> We call that `rate`, because it's our discount rate.

> *Important:* If the user typed something like `"abc"` by mistake, this line will fail — and we’ll catch that in the `except` block below.

---

#### `for item in cart:`

> We now loop through each product that’s currently in the **shopping cart**.

> Think of it like saying:
>
> "Okay, let's go through every item in the cart one by one."

---

#### `item['price'] *= (1 - rate/100)`

> This is the heart of the discount calculation. Let’s break it down:

> * `rate / 100` turns `10` into `0.10` (10%)
>
> * `1 - rate/100` gives us `0.90` (which means 90% of the original price — the discounted amount)
>
> * `item['price'] *= ...` means we're **updating the item's price** by multiplying it with the discounted rate.

> So if the item originally cost Rs. 100 and the user entered 10%, the new price becomes Rs. 90.

---

#### `refresh_cart()`

> After all the prices are updated, we call `refresh_cart()` to **update the cart display on the screen** — so the user can immediately see the new discounted prices and the new total.

---

#### `except ValueError:`

> This part runs **only if there was a problem** in converting the discount to a number.

> For example: if the user typed `abc` or left the box empty, `float(discount)` would crash — and this block will handle that safely.

---

#### `messagebox.showerror("Invalid Input", "Enter a valid discount %")`

> We show a red error popup saying:
>
> **"Invalid Input"** (as the title)
>
> **"Enter a valid discount %"** (as the message)

> It’s a polite way of saying:
>
> "Hey, we couldn’t understand what you typed — please enter a number like 10, 15, etc."

````python
def apply_discount():
    discount = discount_entry.get()
    try:
        rate = float(discount)
        for item in cart:
            item['price'] *= (1 - rate/100)
        refresh_cart()
    except ValueError:
        messagebox.showerror("Invalid Input", "Enter a valid discount %")
````

>
> * **Message:** "Your total bill is Rs. xx.xx. Thank you for shopping!"

> `f"Rs. {total_amount:.2f}"` formats the total to **two decimal places** — so it looks neat, like `Rs. 75.50`.

> We also add a polite "Thank you" — because good apps talk nicely to users

---

#### `cart.clear()`

> Now that the purchase is done, we **empty the cart**.

> Think of this as the cart being reset — like when your basket is cleared after placing an order.

---

#### `refresh_cart()`

> Finally, we update the screen using `refresh_cart()`.

> This will **clear out the displayed cart list and reset the total to Rs. 0.00** — showing the cart is now empty and ready for the next shopping round.

````python
def checkout():
    if not cart:
        messagebox.showinfo("Checkout", "Cart is empty. Please add items to checkout.")
        return
    total_amount = calculate_total()
    messagebox.showinfo("Checkout Complete", f"Your total bill is Rs. {total_amount:.2f}.\nThank you for shopping!")
    cart.clear()
    refresh_cart()
````
