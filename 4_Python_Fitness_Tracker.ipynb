### Load Fitness Data

> This function:

1. Reads a CSV file with fitness data.
2. Separates the numbers (Steps, Sleep, Calories) and dates.
3. Converts the data into formats that are easy to work with.
4. Handles any errors so the program doesn’t break.

---

#### **`def load_fitness_data(file_path):`**

*“We are defining a function named `load_fitness_data`. Think of a function as a small machine or a mini-program. You give it something (in this case, a file path), and it does some work and gives you a result.”*

* `def` is short for *define a function*.
* `load_fitness_data` is the name we’re giving to our function.
* `(file_path)` means this function needs you to give it the location of a file (probably a CSV with fitness data).

---

#### **`try:`**

*“We are going to try running some code. If everything works, great! But if something goes wrong, we’ll catch the error and handle it without crashing the program.”*

This is called **error handling**. You’ll see how in the `except` block.

---

#### **`df = pd.read_csv(file_path)`**

*“Now we’re using Pandas (remember `pd`) to read the CSV file from the path we gave. The data from the file is stored in something called a DataFrame, which is like a table in memory.”*

* `pd.read_csv()` is the function that reads a `.csv` file.
* `df` is the name we’re using to store that table of data.

---

#### **`data = df[["Steps", "SleepHours", "CaloriesBurned"]].to_numpy()`**

*“Here, we are selecting only 3 columns from the table: Steps, SleepHours, and CaloriesBurned. Then we convert that mini table into a NumPy array. Why? Because NumPy arrays are super efficient for calculations.”*

* `df[["col1", "col2"]]` is how we choose specific columns.
* `.to_numpy()` turns the table into a NumPy array.

---

#### **`dates = df["Date"].tolist()`**

*“From the same table, we are pulling out the Date column and converting it into a regular Python list. This makes it easier to print or loop through later.”*

---

#### **`return data, dates`**

*“At this point, our function has done its job: it has loaded the data, split it into numbers and dates, and now it’s sending those two things back to whoever called the function.”*

---

#### **`except Exception as e: ...`**

*“Now, if something goes wrong while loading the file — maybe the file doesn’t exist, or the format is wrong — we don’t want the program to crash. So we print a friendly error message and return `None, None` to show that loading failed.”*

* `Exception as e` means "catch the error and call it `e`"
* `print("Error loading file:", e)` shows what went wrong
* `return None, None` gives a safe fallback so the program can continue

````python
def load_fitness_data(file_path):
    try:
        df = pd.read_csv(file_path)
        data = df[["Steps", "SleepHours", "CaloriesBurned"]].to_numpy()
        dates = df["Date"].tolist()
        return data, dates
    except Exception as e:
        print("Error loading file:", e)
        return None, None
````

### Analyze Week

> This function takes a big dataset, picks 1 week based on the number you give, calculates the averages, prints them nicely, and gives the week’s data back.

---

#### **`def analyze_week(data, week_num):`**

"We are defining a function called `analyze_week`. A function is like a small calculator or helper that we can use whenever we need.

* It takes **two inputs**:

  * `data` → this is your big dataset (probably a NumPy array of fitness data: steps, sleep hours, and calories).
  * `week_num` → this tells us **which week** we want to analyze (Week 1, Week 2, etc.).

So in short, we’re saying: 'Hey Python, look at Week X’s data and tell me the average steps, sleep, and calories.'"

---

#### **`start = (week_num - 1) * 7`**

"Let’s say I want to analyze **Week 2**. How do I know where Week 2 starts in the data?

* Every week has 7 days.
* If it's Week 1 → starts at position 0.
* Week 2 → starts at 7 (because 1 full week = 7 days).

So we use this formula: `(week number - 1) × 7`
This gives us the **starting index** for the week."

---

#### **`end = start + 7`**

"Now that we know where the week starts, we just add 7 to get the end index. So if we started at 7 (Week 2), this gives us 7 + 7 = 14.

Now we know the data we want is from **position 7 to 14** (not including 14)."

---

#### **`week = data[start:end]`**

"This line **cuts out just one week’s data** from the full dataset. We’re slicing the array — like cutting a slice from a cake.

* `data[start:end]` gives us 7 rows — one for each day of the week.
* Now `week` contains 7 days of fitness info (steps, sleep, calories)."

---

#### **`avg = np.mean(week, axis=0)`**

"Now we want the average — like the weekly report card.

* `np.mean()` means: calculate the average.
* `axis=0` tells Python: go **column by column**.

  * So it calculates:

    * average steps,
    * average sleep hours,
    * average calories burned for the week.

We store those 3 numbers in `avg`."

---

#### **Print Command**

```python
print(f"\nWeek {week_num} Avg -> Steps: {avg[0]:.0f}, Sleep: {avg[1]:.1f} hrs, Calories: {avg[2]:.0f}")
```

"This prints out the results nicely.

* It says: 'Hey, here’s the average for Week 1 (or 2, or whatever).'
* We use `f""` formatting to plug in values like a sentence.

  * `avg[0]` → average steps (no decimal).
  * `avg[1]` → average sleep (1 decimal place).
  * `avg[2]` → average calories (no decimal)."

Sample Output:

```
Week 2 Avg -> Steps: 8432, Sleep: 7.4 hrs, Calories: 2110
```

---

#### **`return week`**

"Finally, we **send back the 7-day data** for that week. Why? Maybe the main program wants to do more analysis later — like plotting a chart or checking low-activity days."

````python
def analyze_week(data, week_num):
    start = (week_num - 1) * 7
    end = start + 7
    week = data[start:end]
    avg = np.mean(week, axis=0)
    print(f"\nWeek {week_num} Avg -> Steps: {avg[0]:.0f}, Sleep: {avg[1]:.1f} hrs, Calories: {avg[2]:.0f}")
    return week
````

### Get Low Activity Days

> This function looks through your entire fitness history, checks which days you were inactive, and prints a report for each of those days.

It’s like a **health assistant** checking your records and pointing out:
"Hey! You were low on activity on these days. Take care!"

---

#### **`def get_low_activity_days(data, dates):`**

"Here we’re starting a **function** — a reusable mini-program.

* The name is `get_low_activity_days`, which gives us a clue: we’re going to find days where someone didn’t walk much, didn’t sleep well, or didn’t burn enough calories.
* It takes **two things as input**:

  1. `data` – this is our NumPy array with all the fitness numbers: Steps, Sleep Hours, and Calories.
  2. `dates` – this is the list of dates that matches with each row in `data`.

So we’ll go row by row and check each day’s values."

---

#### **`low_steps = data[:, 0] < 6000`**

"This line is checking **which days had fewer than 6000 steps**.

* `data[:, 0]` means: look at all rows (`:`) but only the first column (which is Steps).
* `< 6000` means we are checking: ‘Was it less than 6000?’

This gives us a **Boolean array** — a list of True/False values.
If True, that means low activity that day in steps."

---

#### **`poor_sleep = data[:, 1] < 6`**

"Same logic here, but now we’re checking for **bad sleep**.

* `data[:, 1]` is the second column — Sleep Hours.
* `< 6` means we are marking all the days with less than 6 hours of sleep.

Again, the result is a list of True/False values."

---

#### **`low_calories = data[:, 2] < 1800`**

"And here, we are checking for days where the person **burned less than 1800 calories**.

* `data[:, 2]` grabs the third column — Calories Burned.
* `< 1800` helps us flag those ‘lazy’ or low-effort days."

---

#### **`print("\nDays with low steps, sleep, or calories:")`**

"This just prints a nice heading before we start showing results."

---

#### **`for i in range(len(data)):`**

"We are now starting a **loop** — we’re going to check each row (each day) one by one.

* `len(data)` gives us the number of days.
* `i` will be 0 for Day 1, 1 for Day 2, and so on."

---

#### **`if low_steps[i] or poor_sleep[i] or low_calories[i]:`**

"This line checks if **any one** of the three problems happened on that day:

* Was the step count low?
* Was sleep poor?
* Were calories too low?

If **any of them** is True, we will go inside the `if` block."

---

#### **`print`**

```python
print(f"{dates[i]} - Steps: {data[i,0]}, Sleep: {data[i,1]}, Calories: {data[i,2]}")
```

"If the condition is True (which means it's a low-activity day), we print:

* the date,
* how many steps were taken,
* how many hours the person slept,
* how many calories they burned.

We use `f""` formatting to plug in values neatly."

````python
#Conditional Filtering
def get_low_activity_days(data, dates):
    low_steps = data[:, 0] < 6000
    poor_sleep = data[:, 1] < 6
    low_calories = data[:, 2] < 1800
    print("\nDays with low steps, sleep, or calories:")
    for i in range(len(data)):
        if low_steps[i] or poor_sleep[i] or low_calories[i]:
            print(f"{dates[i]} - Steps: {data[i,0]}, Sleep: {data[i,1]}, Calories: {data[i,2]}")
````

### Fitness Score

> This function takes one day’s fitness data and gives you a score out of 1 based on how well you walked, slept, and burned calories — with walking having slightly more importance.

It's like your personal fitness report card for the day!

---

#### **`def fitness_score(day):`**

"Here we are **defining a function** called `fitness_score`.

* A function is like a mini machine: we give it something, and it gives something back.
* This one takes **one input**: `day`. Think of `day` as a small bundle that contains fitness data from one day — most likely a list or array of three things:

  1. Steps walked,
  2. Hours slept,
  3. Calories burned.

We’ll calculate a fitness **score** from this day’s data."

---

#### **`steps, sleep, cal = day`**

"This is a **shortcut for unpacking**.

Imagine `day = [8450, 7.5, 1950]`.
This line splits that list into three separate variables:

* `steps = 8450`
* `sleep = 7.5`
* `cal = 1950`

It’s like taking items out of a box and placing them in clearly labeled jars."

---

#### **`return`**

```python
return 0.4 * (steps / 10000) + 0.3 * (sleep / 8) + 0.3 * (cal / 2200)
```

"This line calculates the **fitness score** using a simple formula. Let me explain the logic behind it:

* `steps / 10000`: this checks how close the person was to the ideal 10,000 steps.
* `sleep / 8`: this checks how close their sleep was to 8 hours (which is usually considered healthy).
* `cal / 2200`: this checks how close they came to burning 2200 calories, which could be a target.

Then we apply **weights** to each of these components:

* **40% weight to steps** (that’s `0.4 * (steps/10000)`)
* **30% to sleep**
* **30% to calories**

Finally, all three are **added together** to give us a final number — the **fitness score for the day**.

If all targets are perfectly hit, the score will be:
`0.4 + 0.3 + 0.3 = 1.0` → that’s a **perfect day**!"

````python
#Fitness Score
def fitness_score(day):
    steps, sleep, cal = day
    return 0.4 * (steps / 10000) + 0.3 * (sleep / 8) + 0.3 * (cal / 2200)
````

### Fitness for each day

> This function takes fitness data for many days, calculates a daily fitness score for each day using a formula we already defined, and returns a clean NumPy array of all those scores.

It’s like turning a whole month's fitness log into a neat **score sheet**, where each score reflects how healthy the person was that day.

---

#### **`def score_all_days(data):`**

“We are **defining a new function** here. The name of the function is `score_all_days`, and it does exactly what it sounds like: it gives a fitness score for **each day** in the dataset.

* The input is called `data`, which is most likely a **NumPy array**.
* Each row in this `data` array represents one day.
* Each row probably contains **three values**: Steps, Sleep Hours, and Calories Burned — just like in our earlier function `fitness_score(day)`."

So this function is going to use that earlier function — but now apply it to **many days**, not just one.

---

#### **`return`**

```python
return np.array([fitness_score(day) for day in data])
```

“Now this line is where all the action happens. Let’s unpack it piece by piece.”

#### First part: `[fitness_score(day) for day in data]`

* This is called a **list comprehension**.
  It’s a fancy Python shortcut that means:

  > “Go through each `day` in `data`, and for each one, calculate `fitness_score(day)`.”

Think of `data` like a table with 14 rows (for 14 days of fitness tracking).
This line goes row by row, sending each row to the `fitness_score()` function, and collecting the results in a list.

#### Second part: `np.array(...)`

* Once the list of scores is ready, we wrap it with `np.array()` to turn it into a **NumPy array**.

Why? Because working with arrays is faster and better for math or plotting.

````python
def score_all_days(data):
    return np.array([fitness_score(day) for day in data])
````

### Generate Summary

> This function takes in fitness data and date labels. Then it:
>
> * Finds the most and least active days based on steps.
> * Calculates fitness scores for all days.
> * Identifies the top 3 best overall fitness days based on combined score.

---

#### **`def generate_summary(data, dates):`**

"This line **defines a function** called `generate_summary`. Think of this function like a mini-report generator.

* It takes two inputs:

  * `data`: a NumPy array containing daily fitness data (probably columns like Steps, Sleep, Calories).
  * `dates`: a list of dates corresponding to each row of the data.

So this function will look at the numbers, compare days, and help us summarize how the person did."

---

#### **`best_day_idx = np.argmax(data[:, 0])`**

"This line is figuring out **which day had the highest step count**.

* `data[:, 0]` means: look at all rows (`:`) and only the first column (which represents **Steps**).
* `np.argmax(...)` returns the **index** (or position) of the **highest value** in that list.

Example: If the most steps were on Day 5, this will return `4` (because Python starts counting from 0).

We’ll use this index to look up the best date later."

---

#### **`worst_day_idx = np.argmin(data[:, 0])`**

"This does the **opposite** of the previous line. It finds the **index of the day with the lowest step count**.

* `np.argmin(...)` returns the index of the smallest value — the day when the person was least active.

So by now, we’ve got the most and least active days."

---

#### **`scores = score_all_days(data)`**

"Here we are calling a function we defined earlier — `score_all_days()`.

* It will go through each row in `data` and calculate a **fitness score** based on steps, sleep, and calories.
* Then it returns a NumPy array of those scores.

Think of it as turning the entire fitness log into a list of numbers between 0 and 1 — higher means healthier."

---

#### **`top3_idx = np.argsort(scores)[-3:][::-1]`**

"This is a cool trick in Python to find the **top 3 best days** based on fitness score.

Let’s break it into parts:

* `np.argsort(scores)` sorts the indexes based on score — from **lowest to highest**.
* `[-3:]` takes the **last 3 items** from that sorted list — which are the 3 **highest scores**.
* `[::-1]` reverses them — so the **best day comes first**, then second best, and third best.

Result: `top3_idx` is a list of the 3 best-performing days (by index) in descending order."

---

#### Print Best Day Based On Steps Walked

```python
print(f"\nBest Step Day: {dates[best_day_idx]} ({data[best_day_idx, 0]} steps)")
```

* `\n` is just to give some space before the output (starts on a new line).
* `dates[best_day_idx]` picks the **date** of the best day — remember, `best_day_idx` was calculated earlier using `np.argmax(...)`.
* `data[best_day_idx, 0]` means: go to the row `best_day_idx` in the data and pick column 0 (which is **steps**).

So the printed message would look like:
**Best Step Day: 2025-05-12 (14200 steps)**
Nice and clear!"

---

#### Print Worst Day

```python
print(f"Worst Step Day: {dates[worst_day_idx]} ({data[worst_day_idx, 0]} steps)")
```

"Same idea as the previous line — but now it's printing the **worst day** based on steps.

* `worst_day_idx` gives us the index of the day with the lowest number of steps.
* We’re showing both the date and the step count."

Example output:
**Worst Step Day: 2025-05-06 (4250 steps)**

---

#### Print Fitness Score

```python
print("\nTop 3 Fitness Days:")
```

"This is a heading to introduce the top 3 best fitness days overall — based on the **fitness score** (not just steps)."

* The `\n` moves to a new line again.
* The next lines will list the best days, one by one."

---

#### Print Top 3 Fitness Days

```python
for idx in top3_idx:
    print(f"{dates[idx]} - Score: {scores[idx]:.2f}")
```

"This is a **loop** — we are going through each of the top 3 best fitness days:

* `top3_idx` is a list of the indexes of the 3 highest scoring days (we calculated that earlier using sorting).
* For each index (`idx`), we:

  * Look up the date from the `dates` list: `dates[idx]`
  * Get the fitness score from the `scores` list: `scores[idx]`
  * Use `:.2f` to round the score to **2 decimal places** for neatness.

Example output:

```
2025-05-03 - Score: 0.91
2025-05-15 - Score: 0.88
2025-05-10 - Score: 0.86
```

So this gives a quick summary of their most productive, healthiest days.


````python
def generate_summary(data, dates):
    best_day_idx = np.argmax(data[:, 0])
    worst_day_idx = np.argmin(data[:, 0])
    scores = score_all_days(data)
    top3_idx = np.argsort(scores)[-3:][::-1]

    print(f"\nBest Step Day: {dates[best_day_idx]} ({data[best_day_idx, 0]} steps)")
    print(f"Worst Step Day: {dates[worst_day_idx]} ({data[worst_day_idx, 0]} steps)")
    print("\nTop 3 Fitness Days:")
    for idx in top3_idx:
        print(f"{dates[idx]} - Score: {scores[idx]:.2f}")
````

### Save Summary to File

> This function opens a text file and writes a short summary:
>
> * A title
> * Number of active days (where steps were 10,000+)
> * Average sleep hours
> * Average calories burned

---

#### `def save_summary_to_file(data, dates, filename="fitness_summary.txt"):`

“We’re **defining a function** called `save_summary_to_file`. This function does what its name suggests — it **writes a fitness summary to a text file**.

* `data`: this is your fitness dataset (steps, sleep, calories).
* `dates`: the list of dates (though in this function, we won’t actually use it).
* `filename`: the name of the file where we’ll **save the summary**. It has a default name `"fitness_summary.txt"`.

So even if you don’t give a file name while calling this function, it’ll automatically save to that default file.”

“This line opens a **new file** to write into:

* `as f` lets us use `f` as a short name to refer to the file throughout this block.

Python automatically closes the file after this block is finished, so it's safe and clean.”

---

#### `f.write("Fitness Tracker Summary Report\n\n")`

“We’re now **writing the title of the report** into the file.

* `\n\n` adds two blank lines for spacing to make the file more readable.
* After this line, if you open the text file, it will start with:

  ```
  Fitness Tracker Summary Report
  ```

Nice and professional-looking.”

---

#### `active_days = np.sum(data[:, 0] >= 10000)`

“This line **calculates how many days had 10,000 or more steps** — which we’ll call active days.

* `data[:, 0]` gets the **Steps column**.
* `data[:, 0] >= 10000` creates a list of `True` and `False` values — one for each day.
* `np.sum(...)` counts how many `True` values there are — that’s our number of active days!”

---

#### `f.write(f"Total Active Days (Steps > 10,000): {active_days}\n")`

“Now we write that count into the file, in a sentence format that’s easy to read.”

Sample output in the file might be:
**Total Active Days (Steps > 10,000): 9**

---

#### `f.write(f"Average Sleep: {np.mean(data[:, 1]):.2f} hrs\n")`

“This line writes the **average sleep hours** into the file.

* `data[:, 1]` gives the **Sleep Hours column**.
* `np.mean(...)` calculates the average.
* `:.2f` rounds it to **2 decimal places**.”

Example output:

**Average Sleep: 7.25 hrs**

---

#### `f.write(f"Average Calories Burned: {np.mean(data[:, 2]):.0f} kcal\n")`

“Same idea as above — but this time we’re calculating and saving the **average calories burned**.

* `data[:, 2]` is the Calories column.
* `:.0f` means we round to **no decimal places**, because calories are usually reported as whole numbers.”

Example output:

**Average Calories Burned: 2123 kcal**

---

#### `print(f"\nSummary saved to {filename}")`

“Finally, we print a message on the screen (not in the file) to tell the user that the summary has been saved — and show the filename.”

On screen:

```
Summary saved to fitness_summary.txt
```

````python
def save_summary_to_file(data, dates, filename="fitness_summary.txt"):
    with open(filename, "w") as f:
        f.write("Fitness Tracker Summary Report\n\n")
        active_days = np.sum(data[:, 0] >= 10000)
        f.write(f"Total Active Days (Steps > 10,000): {active_days}\n")
        f.write(f"Average Sleep: {np.mean(data[:, 1]):.2f} hrs\n")
        f.write(f"Average Calories Burned: {np.mean(data[:, 2]):.0f} kcal\n")
    print(f"\nSummary saved to {filename}")
````
